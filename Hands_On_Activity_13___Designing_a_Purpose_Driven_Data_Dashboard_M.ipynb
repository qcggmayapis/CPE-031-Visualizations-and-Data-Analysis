**3. Discussions:**

A well-designed data dashboard is more than a collection of charts — it’s a purposeful storytelling tool that connects data to decisions.

Understanding the Purpose of the Data Dashboard:
Every dashboard must have a clear objective — whether it’s to monitor performance, analyze patterns, or inform strategic decisions. The design should align with this goal, focusing only on the most relevant data and metrics.

Considering the Needs of the Data Dashboard’s Users:
Dashboards should be designed with the user in mind. A manager may need high-level KPIs, while an analyst may need granular, filterable data. Knowing the user’s goals, skills, and workflow ensures that the dashboard supports, rather than overwhelms, their decision-making.

Data Dashboard Engineering:
Engineering a dashboard involves organizing visual elements efficiently, ensuring smooth interactivity, and maintaining data accuracy. It also includes performance considerations (like handling large datasets), responsive design, and consistent color schemes.

**4. Procedures:**

Step 1: Install and Import Required Libraries

```python
!pip install plotly pandas
import pandas as pd
import numpy as np
import plotly.express as px
```

Step 2: Create a Sample Dataset

We’ll simulate student performance data to design an academic performance dashboard.

```python
# Create dataset
np.random.seed(10)
students = [f"Student {i}" for i in range(1, 21)]
subjects = ["Math", "Science", "English", "Programming", "Data Analytics"]

data = {
    "Student": np.random.choice(students, 100),
    "Subject": np.random.choice(subjects, 100),
    "Score": np.random.randint(60, 100, 100),
    "Attendance (%)": np.random.randint(70, 100, 100)
}

df = pd.DataFrame(data)
df.head()
```

Step 3: Design Dashboard Components

```python
# Average Score per Subject
fig1 = px.bar(df.groupby("Subject", as_index=False)["Score"].mean(),
              x="Subject", y="Score", title="Average Score per Subject",
              color="Score", color_continuous_scale="Blues")

# Attendance vs. Score Scatter
fig2 = px.scatter(df, x="Attendance (%)", y="Score", color="Subject",
                  title="Attendance vs. Performance", hover_data=["Student"])

# Score Distribution
fig3 = px.box(df, x="Subject", y="Score", title="Score Distribution by Subject",
              color="Subject")

fig1.show()
fig2.show()
fig3.show()
```

Step 4: Combine into One Interactive Dashboard

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Aggregated data
avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
avg_attendance = df.groupby("Subject", as_index=False)["Attendance (%)"].mean()

# Dashboard layout
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Score per Subject", "Attendance vs. Score",
                                    "Score Distribution", "Average Attendance per Subject"),
                    specs=[[{"type": "bar"}, {"type": "scatter"}],
                           [{"type": "box"}, {"type": "bar"}]])

fig.add_trace(go.Bar(x=avg_scores["Subject"], y=avg_scores["Score"], name="Average Score"), row=1, col=1)
fig.add_trace(go.Scatter(x=df["Attendance (%)"], y=df["Score"], mode="markers", name="Scatter Plot"), row=1, col=2)
fig.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Score Distribution"), row=2, col=1)
fig.add_trace(go.Bar(x=avg_attendance["Subject"], y=avg_attendance["Attendance (%)"], name="Average Attendance"), row=2, col=2)

fig.update_layout(height=700, width=950, title_text="🎓 Student Performance Dashboard")
fig.show()
```

**Task 2:** Design for User Needs

Modify the dashboard to meet specific user needs:

If your user is a teacher, highlight underperforming students.

If your user is a student, highlight their personal progress vs. class average.

If your user is a department head, summarize subject-level averages only.

```python
selected_student = "Student 5" # Choose a specific student

# Calculate selected student's average score per subject
student_avg_scores = df[df["Student"] == selected_student].groupby("Subject", as_index=False)["Score"].mean()
student_avg_scores.rename(columns={"Score": f"{selected_student} Average Score"}, inplace=True)

# Calculate overall class average score per subject
class_avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
class_avg_scores.rename(columns={"Score": "Class Average Score"}, inplace=True)

# Merge for comparison
comparison_df = pd.merge(class_avg_scores, student_avg_scores, on="Subject", how="left").fillna(0)
print("Comparison DataFrame head:")
print(comparison_df.head())

import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px

# --- Create charts for Student ---

# Pastel colors for bars and scatter points
pastel_colors_discrete = px.colors.qualitative.Pastel

# 1. Comparison Bar Chart: Student Average Score vs. Class Average Score
fig_student_vs_class_avg = go.Figure(
    data=[go.Bar(name='Class Average', x=comparison_df['Subject'], y=comparison_df['Class Average Score'], marker_color=pastel_colors_discrete[0]),
          go.Bar(name=f'{selected_student} Average', x=comparison_df['Subject'], y=comparison_df[f'{selected_student} Average Score'], marker_color=pastel_colors_discrete[1])]
)
fig_student_vs_class_avg.update_layout(barmode='group', title_text='Your Average Score vs. Class Average Score per Subject')

# Filter for the selected student's data
student_data = df[df['Student'] == selected_student]

# 2. Student's Individual Scores (Bar Chart)
fig_student_individual_scores = px.bar(student_data,
                                         x='Subject',
                                         y='Score',
                                         title=f'{selected_student}\'s Individual Scores per Subject',
                                         hover_data=['Attendance (%)'],
                                         color='Score', # Color by score magnitude
                                         color_continuous_scale=px.colors.sequential.Plotly3)

# 3. Student's Attendance vs. Score (Scatter Plot)
fig_student_attendance_vs_score = px.scatter(student_data,
                                              x='Attendance (%)',
                                              y='Score',
                                              title=f'{selected_student}\'s Attendance vs. Score',
                                              hover_data=['Subject'],
                                              color='Subject',
                                              color_discrete_sequence=pastel_colors_discrete) # Use pastel

# Dashboard layout for Student Persona
fig_student_dashboard = make_subplots(rows=2, cols=2,
                                        subplot_titles=("Your Average vs. Class Average Scores",
                                                        "Your Individual Scores per Subject",
                                                        "Your Attendance vs. Score",
                                                        "Overall Performance Distribution"),
                                        specs=[[{"type": "bar"}, {"type": "bar"}],
                                               [{"type": "scatter"}, {"type": "box"}]])

# Add traces to the student dashboard
fig_student_dashboard.add_trace(fig_student_vs_class_avg['data'][0], row=1, col=1) # Class Average Bar
fig_student_dashboard.add_trace(fig_student_vs_class_avg['data'][1], row=1, col=1) # Student Average Bar
fig_student_dashboard.add_trace(fig_student_individual_scores['data'][0], row=1, col=2)
fig_student_dashboard.add_trace(fig_student_attendance_vs_score['data'][0], row=2, col=1)

# Re-using the class-wide score distribution as context.
fig_student_dashboard.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Overall Score Distribution", showlegend=False, marker_color=pastel_colors_discrete[2]),
                                 row=2, col=2)

# Update layout
fig_student_dashboard.update_layout(height=800, width=1000,
                                    title_text=f"🎓 Student Performance Dashboard ({selected_student}'s View)",
                                    barmode='group')

fig_student_dashboard.show()
```

**5. Supplementary Activity:**

Design Your Own Dashboard: “Smart Campus Energy Monitor”
Create a dashboard that shows the energy usage of different campus areas (e.g., laboratories, offices, classrooms) over time.

1. Simulate or use sample data (Energy consumption per hour or per day).

2. Include at least three charts (bar, line, pie).

3. Ensure your design reflects a clear purpose and target user (e.g., facilities manager).

4. Write a short explanation of your engineering decisions (why you chose certain colors, layouts, or interactivity).

```python
import pandas as pd
import numpy as np

# --- Define campus areas ---
campus_areas = ['Laboratories', 'Offices', 'Classrooms', 'Library', 'Athletic Center']

# --- Generate timestamps (hourly from Oct 2 to Nov 2, 2025) ---
start_date = '2025-10-02 00:00:00'
end_date = '2025-11-02 23:00:00'
timestamps = pd.date_range(start=start_date, end=end_date, freq='h')

# --- Create random energy consumption data ---
energy_data = []
for ts in timestamps:
    for area in campus_areas:
        if area == 'Laboratories':
            consumption = np.random.uniform(20, 150)
        elif area == 'Offices':
            consumption = np.random.uniform(5, 40)
        elif area == 'Classrooms':
            consumption = np.random.uniform(10, 60)
        elif area == 'Library':
            consumption = np.random.uniform(15, 80)
        else:  # Athletic Center
            consumption = np.random.uniform(8, 70)

        energy_data.append({
            'Timestamp': ts,
            'Campus Area': area,
            'Energy Consumption (kWh)': round(consumption, 2)
        })

# --- Convert to DataFrame ---
energy_df = pd.DataFrame(energy_data)

# --- Summarize total energy consumption ---
total_energy_per_area = energy_df.groupby('Campus Area')['Energy Consumption (kWh)'].sum().reset_index()

# --- Show output ---
print(f"{len(energy_df)} energy consumption records.")
print("\nSample of energy data:")
print(energy_df.head())

print("\nTotal energy consumption per campus area:")
print(total_energy_per_area)

facility_colors = {
    'Laboratories': '#2E86AB',
    'Offices': '#58A55C',
    'Classrooms': '#F4B400',
    'Library': '#A569BD',
    'Athletic Center': '#E67E22'
}
```

```python
import plotly.express as px

# --- Bar Chart ---
fig_bar = px.bar(
    total_energy_per_area, x='Campus Area', y='Energy Consumption (kWh)',
    title='Total Energy Consumption per Campus Area (for facilities manager)',
    color='Campus Area', color_discrete_map=facility_colors
)
fig_bar.update_layout(template='simple_white', title_font_size=20, font=dict(size=14), showlegend=False)
fig_bar.show()
```

```python
import plotly.express as px

# --- Line Chart ---
fig_line = px.line(
    energy_df, x='Timestamp', y='Energy Consumption (kWh)',
    color='Campus Area', color_discrete_map=facility_colors,
    title='Energy Consumption Trends Over Time by Campus Area (for facilities manager)'
)
fig_line.update_layout(template='simple_white', title_font_size=20, font=dict(size=14))
fig_line.show()
```

```python
import plotly.express as px

fig_pie = px.pie(
    total_energy_per_area,
    values='Energy Consumption (kWh)',
    names='Campus Area',
    title='Proportional Energy Distribution Across Campus Areas (for facilities manager)',
    color='Campus Area',
    color_discrete_map=facility_colors
)
fig_pie.update_traces(textinfo='percent+label', pull=[0.02]*len(total_energy_per_area))
fig_pie.update_layout(template='simple_white', title_font_size=20, font=dict(size=14))
fig_pie.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px

campus_areas = ['Laboratories', 'Offices', 'Classrooms', 'Library', 'Athletic Center']

# Generate hourly timestamps for one month
start_date = '2025-10-02 00:00:00'
end_date = '2025-11-02 23:00:00'
timestamps = pd.date_range(start=start_date, end=end_date, freq='h')

# Create energy consumption dataset
energy_data = []
for ts in timestamps:
    for area in campus_areas:
        if area == 'Laboratories':
            consumption = np.random.uniform(20, 150)
        elif area == 'Offices':
            consumption = np.random.uniform(5, 40)
        elif area == 'Classrooms':
            consumption = np.random.uniform(10, 60)
        elif area == 'Library':
            consumption = np.random.uniform(15, 80)
        else:  # Athletic Center
            consumption = np.random.uniform(8, 70)

        energy_data.append({
            'Timestamp': ts,
            'Campus Area': area,
            'Energy Consumption (kWh)': round(consumption, 2)
        })

# Create DataFrame
energy_df = pd.DataFrame(energy_data)

# Total energy per area
total_energy_per_area = energy_df.groupby('Campus Area')['Energy Consumption (kWh)'].sum().reset_index()

# Display outputs
print(f"✅ Generated {len(energy_df)} energy consumption records.\n")
print("🔹 Sample of energy data:")
print(energy_df.head())
print("\n🔹 Total energy consumption per campus area:")
print(total_energy_per_area)

facility_colors = {
    'Laboratories': '#2E86AB',
    'Offices': '#58A55C',
    'Classrooms': '#F4B400',
    'Library': '#A569BD',
    'Athletic Center': '#E67E22'
}

# --- Bar Chart ---
fig_bar = px.bar(
    total_energy_per_area,
    x='Campus Area',
    y='Energy Consumption (kWh)',
    title='Total Energy Consumption per Campus Area (for facilities manager)',
    color='Campus Area',
    color_discrete_map=facility_colors,
    labels={'Energy Consumption (kWh)': 'Total Energy Consumption (kWh)'}
)
fig_bar.update_layout(template='simple_white', title_font_size=20, font=dict(size=14), showlegend=False)
fig_bar.show()

# --- Line Chart ---
fig_line = px.line(
    energy_df,
    x='Timestamp',
    y='Energy Consumption (kWh)',
    color='Campus Area',
    title='Energy Consumption Trends Over Time by Campus Area (for facilities manager)',
    color_discrete_map=facility_colors,
    labels={'Energy Consumption (kWh)': 'Energy Consumption (kWh)', 'Timestamp': 'Time'}
)
fig_line.update_layout(template='simple_white', title_font_size=20, font=dict(size=14))
fig_line.show()

# --- Pie Chart ---
fig_pie = px.pie(
    total_energy_per_area,
    values='Energy Consumption (kWh)',
    names='Campus Area',
    title='Proportional Energy Distribution Across Campus Areas (for facilities manager)',
    color='Campus Area',
    color_discrete_map=facility_colors
)
fig_pie.update_traces(textinfo='percent+label', pull=[0.02]*len(total_energy_per_area))
fig_pie.update_layout(template='simple_white', title_font_size=20, font=dict(size=14))
fig_pie.show()
```

This visualizations was designed with professionalism and clarity to support a facilities manager's decision-making. These colors and charts make the energy data easy to understand, with blue for laboratories, green for offices, gold for classrooms, purple for the library, and orange for the athletic center. In the bar chart, it shows what area uses the most energy, line chart shows how the energy changes over time, and the pie chart shows the percentage of energy that was being used by each area.
