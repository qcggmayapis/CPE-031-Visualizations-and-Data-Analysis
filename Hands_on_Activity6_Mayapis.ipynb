```python
import matplotlib.pyplot as plt

# The regions of the OrderDate
region_counts = df['Region'].value_counts()
#Filter the DataFrame for the specific dat
df_filtered_date = df[df['OrderDate'] == '7/21/22']

# Count the occurrences of each region for the filtered data
region_counts = df_filtered_date['Region'].value_counts()

# Plot a bar chart
region_counts.plot.bar(figsize = (8, 6), color = 'skyblue', edgecolor = 'black')

# Add labels and title
plt.title('Regions that made a purchase on 7/21/22')
plt.xlabel('7/21/22')
plt.ylabel('Region')
plt.title('Regions that made a purchase on 7/21/22')
plt.xlabel('Region')
plt.ylabel('Number of Purchases')

# Show chart
plt.show()
```

**Descriptive Analysis:** I used a bar chart here to easily determine how many regions that have purchased in 7/21/22. Where it shows that there's only one region that only purchased an item during that date, and that is the region of Central where it purchases 3 similar items in a same day. That item was pen set and each pen set have 55 units.

```python
df[df['Region'] == 'Central']
```

```python
import matplotlib.pyplot as plt

# Convert the 'Total' column to numeric, handling commas
df['Total'] = df['Total'].astype(str).str.replace(',', '', regex = False)
df['Total'] = pd.to_numeric(df['Total'])

# Group by Rep and sum the Total sales
rep_total_sales = df.groupby('Rep')['Total'].sum()

# Plot bar chart (horizontal bars)
plt.figure(figsize=(10, 8))
rep_total_sales.plot(kind = 'barh', color = 'lightcoral', edgecolor = 'black')

# Add labels and title
plt.title('Total Costs by Each Representative in Central')
plt.xlabel('Total Costs')
plt.ylabel('Representative')

# Show chart
plt.show()
```

**Descriptive Analysis:** Here I used the horizontal graph to differentiate the total costs in each item of the different representative in region Central. The data shows that the Parent have the highest purchased of items in region central, while the lowest cost of items goes to Andrews which it only have below 500.

```python
df[df['Rep'] == 'Jardine']
```

```python
import matplotlib.pyplot as plt

# Filter Jardine's data
jardine_data = df[df['Rep'] == 'Jardine']

# Group by item and sum their total costs
items_costs = jardine_data.groupby('Item')['Total'].sum().sort_values(ascending = False)

# Calculate grand total
grand_total = items_costs.sum()

# Create a bar chart
items_costs.plot.bar(figsize = (8,6), color = 'skyblue', edgecolor = 'black')

# Add Labels and title
plt.title('Total Costs of Items Bought by Jardine')
plt.xlabel('Item')
plt.ylabel('Total Costs')

plt.xticks(rotation = 0, ha = 'right')

for i, value in enumerate(items_costs):
    plt.text(i, value + (value * 0.02), f'${value:.2f}', ha = 'center')

# Show chart
plt.show()
```

**Descriptive Analysis:** I also used the bar chart here to differentiate the total costs of the items that Jardine bought. With bar chart, it's much easier to determine the highest cost item which is the binder and the lowest cost item, the pen set. With this, it will be more easier to know the differences between the items when it comes to costs.
